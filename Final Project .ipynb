* <strong><em> Residual Sugar </em></strong>: Residual sugar represents the amount of sugar remaining after alcoholic fermentation, when yeasts have converted grape sugars into alcohol. The higher the level of residual sugar, the sweeter the wine and the fuller or more viscous it can taste in the mouth.
Managing sugar levels include:
lowering the temperature  to stop fermentation when the wine reaches the desired sweetness.
adding sulfur dioxide to stop fermentation.
adding neutral grape alcohol to stop fermentation, such as in fortified wines like Port.
adding unfermented grape must (the juice before fermentation) to the fermented wine to increase the amount of sugar.

* __Volatile Acidity__: In general "The maximum volatile acidity, calculated as acetic acid and exclusive of sulfur dioxide, is 0.14 g/100 mL for red wine and 0.12 g/100 mL for white wines." This is equivalent to 1.4 and 1.2 g/L acetic acid for red and white wines, respectively

* __Fixed acidity__ : most acid involved with wine or fixed or nonvolatile(do not evaporate readily)

* __Citric acid__: found in small quantities, citric acid can add 'freshness'and flavors to wine 

* __clhorides__: amount of salt in the wine 

* __free_sulfure dioxyde__: the free from so2 exists in equilibrium betwen molecular SO2 and bisulfite ion : it prevent microbial growth and the oxidation of wine

* __total_sulfure_dioxyde__: amount of free and bounds form of SO2 , in low concentration SO2 is mostly indetectable in wine, 2-but at free SO2 concentration over 50 ppm SO2 becomes evident in the taste of wine 

* __density__ : wine density 

* __PH__ : describes how acidic or basic a wine is on a scale 0(very acidic) to 14 14(very basic) most wine are PH betwen 3 and 4 

* __SULPHATES__: a wine additive which can contribute to sulfur dioxyd gas (SO2)levels which act as antimicrobial and antioxydant 

* __Alcahol__: the percent alcahol contient of the wine

* __Quality__: output variable based on sensory data and scaled betwen 0 to 10

```python
# Test Anderson Darling 
import plotly.graph_objs as go
import plotly.figure_factory as ff

import numpy as np
import pandas as pd
import scipy
import chart_studio
import chart_studio.plotly as py
import plotly.graph_objects as go
from scipy.stats import anderson
from plotly.offline import plot

result = anderson(v_blanc['alcohol'])
stat = round(result.statistic, 4)

p = 0
result_mat = []
for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < result.critical_values[i]:
        msg = 'Sample looks Gaussian (fail to reject H0)'
    else:
        msg = 'Sample does not look Gaussian (reject H0)'
    result_mat.append([len(v_blanc[['alcohol']]), stat, sl, cv, msg])

trace = go.Table(
    header=dict(values=['<b>Sample Size</b>', '<b>Statistic</b>', '<b>Significance Level</b>', '<b>Critical Value</b>', '<b>Comment</b>'],
                line = dict(width=0),
                fill = dict(color='rgba(42,63,95,0.8)'),
                align = 'center',
                font = dict(
                    color = '#ffffff',
                    size = 12
                )),
    cells=dict(values=np.array(result_mat).T,
               line = dict(width=0),
               fill = dict(color=[['#EBF0F8', '#ffffff', '#EBF0F8', '#ffffff', '#EBF0F8']]),
               align = 'center',
               height = 40),
    columnwidth=[0.3, 0.25, 0.3, 0.25, 0.5])
layout = dict(
    height=300,
    margin=dict(
        l=5,
        r=5,
        t=30,
        b=0
    )
)
data = [trace]
andar_table = dict(data=data, layout=layout)

plot(andar_table, filename='anderson-darling-table')
```

```python
def call_red(functionToCall):
    print('\nRed Wine\n')
    functionToCall(v_rouge)
def call_white(functionToCall):
    print('\nWhite Wine\n')
    functionToCall(v_blanc)
    
def rm1(red1_set):
    red1_set.columns = [x.strip() for x in red1_set.columns]
call_red(rm1)

def rm2(white_set):
    white_set.columns = [x.strip() for x in white_set.columns]
call_white(rm2)
```
