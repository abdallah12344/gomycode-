```python
scaler = StandardScaler() # standardisation of data 
data_std = scaler.fit_transform(data)

n_clusters = 10

clustering = KMeans(n_clusters=n_clusters,
                    random_state=0
                   )

cluster_labels = clustering.fit_predict(data)

# plot cluster sizes

plt.hist(cluster_labels, bins=range(n_clusters+1))
plt.title ('Customers per Cluster')
plt.xlabel('Cluster')
plt.ylabel('Customers')
plt.show()

```

```python
#Determining Number of Clusters with Elbow Method
ssd= []
cluster_list = range(1, 11)
for i in cluster_list :
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 40)
    kmeans.fit(data_std)
    ssd.append(kmeans.inertia_)
```

```python
plt.plot(cluster_list, ssd)
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('ssd')
plt.show()
```

```python
# the elbow at 5 let's we check with silouhette method 
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = [] 

for n_cluster in range(2, 11):
    silhouette_scores.append( 
        silhouette_score(data_std, KMeans(n_clusters = n_cluster).fit_predict(data_std))) 
    
# Plotting a bar graph to compare the results 
k = [2, 3, 4, 5, 6,7,8,9,10] 
plt.bar(k, silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show()
```

```python
siliuette_list_hierarchical = []

for cluster in range(2,20,2):
    for linkage_method in ['ward', 'average', 'complete']:
        agglomerative = AgglomerativeClustering(linkage=linkage_method, affinity='euclidean',n_clusters=cluster).fit_predict(data_std)
        sil_score = metrics.silhouette_score(data_std, agglomerative, metric='euclidean')
        siliuette_list_hierarchical.append((cluster, sil_score, linkage_method, len(set(agglomerative)) ) )
        
data_hierarchical = pd.DataFrame(siliuette_list_hierarchical, columns=['cluster', 'sil_score','linkage_method', 'number_of_clusters'])
data_hierarchical.sort_values('sil_score', ascending=False)
```

```python
 
kmeans_values=[]

for cluster in range(2,20):
    kmeans = KMeans(n_clusters=cluster, random_state=40).fit_predict(data_std)
    sil_score = metrics.silhouette_score(data_std,kmeans, metric='euclidean')
    print("Silhouette score for {} cluster k-means: {:.3f}".format(cluster,
                                                               metrics.silhouette_score(data_std, kmeans, metric='euclidean')))
    kmeans_values.append((cluster,sil_score))
```
